Sınıflandırma Nedir?: Sınıflandırma, bir veri kümesini belirli sınıflara ayırma işlemidir. Bu sınıflar genellikle etiketler veya temel gerçeklikleri temsil eder.

İkili ve Çok Sınıflı Sınıflandırma: İkili sınıflandırma, veriyi iki sınıfa ayırırken, çok sınıflı sınıflandırma üç veya daha fazla sınıfa ayırır.

Lojistik Regresyon: Lojistik regresyon, doğrusal bir regresyon gibi görünse de, sınıflandırma problemlerini çözmek için kullanılır. Sigmoid işlevi kullanarak sonuçları 0 ile 1 arasında sınırlar ve ikili sınıflandırma için idealdir.

Sınıflandırma Performans Metrikleri: Model performansını değerlendirmek için kullanılan metrikler şunlardır: doğruluk (accuracy), kesinlik (precision), geri çağırma (recall), ve F1 puanı.

Destek Vektör Makineleri (SVM): SVM, sınıflandırma problemlerini çözmek için kullanılan bir başka denetimli öğrenme modelidir. SVM, farklı sınıfları ayıran bir karar sınırı (hiper düzlem) bulur.

Veri Hazırlığı: Veri kümesinin temizlenmesi ve sayısal hale getirilmesi gerekebilir. Etiketler genellikle sayılara dönüştürülür.

Model Eğitme ve Değerlendirme: Veri kümesi tren ve test setlerine ayrılarak bir sınıflandırma modeli eğitilir ve ardından test verisi üzerinde değerlendirilir.

Sonuçlar: Karışıklık matrisi ve sınıflandırma raporu gibi performans metrikleri kullanılarak modelin başarısı değerlendirilir.

```python
import pandas as pd
dataset = pd.read_csv("breast-cancer.csv")
```

\Bu kod, pandas kütüphanesini içe aktarıyor ve "breast-cancer.csv" adlı bir CSV dosyasından veri setini okuyor. Veri seti kanser teşhisi için kullanılacak olan özellikleri içeriyor.



```python
dataset.shape
```

Bu kod, veri setinin boyutunu (satır ve sütun sayısı) gösterir. Bu veri setinde toplam 569 örnek ve 31 özellik bulunmaktadır.

```python
dataset.head()
```

```python
dataset.tail()
```

Bu kod, "diagnosis" sütunundaki sınıfları (M: Malignant, B: Benign) sayısal değerlere dönüştürmek için LabelEncoder kullanır. Bu, sınıflandırma algoritması için giriş verisini hazırlar.

// m=1 b=0 olarak güncellendi


```python
from sklearn.preprocessing import LabelEncoder
labelencoder = LabelEncoder()
dataset["diagnosis"] = labelencoder.fit_transform(dataset["diagnosis"].values)
```

```python
dataset.tail()
```

```python
from sklearn.model_selection import train_test_split
train, test = train_test_split(dataset, test_size=0.3)
```

X_train = train.drop("diagnosis", axis=1): Bu satır, eğitim verilerinin bağımsız değişkenlerini (X_train) hazırlar. train veri çerçevesinden "diagnosis" sütununu çıkartarak (drop) bağımsız değişkenleri elde eder. "diagnosis" sütunu, veri setindeki hedef değişkeni temsil eder, bu nedenle bağımsız değişkenlerin bu sütun dışında olması gerekir.

y_train = train.loc[:, "diagnosis"]: Bu satır, eğitim verilerinin bağımlı değişkenini (y_train) hazırlar. train veri çerçevesindeki "diagnosis" sütununu (loc kullanarak) bağımlı değişken olarak belirler.

```python
X_train = train.drop("diagnosis", axis=1)
y_train = train.loc[:, "diagnosis"]
X_test = test.drop("diagnosis", axis=1)
y_test = test.loc[:, "diagnosis"]
```

 scikit-learn kütüphanesinden Logistic Regression sınıflandırma modelini içe aktarır ve eğitim verileri üzerinde bu modeli eğitir.

```python
from sklearn.linear_model import LogisticRegression
model_1 = LogisticRegression()
model_1.fit(X_train, y_train)
```

```python
predictions = model_1.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix, classification_report
confusion_matrix(y_test, predictions)
print(classification_report(y_test, predictions))
```

```python
from sklearn.svm import LinearSVC
#sınıflandırma modelini içe aktarır
```

```python
model_2 = LinearSVC()
model_2.fit(X_train, y_train)
predictions = model_2.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix, classification_report
confusion_matrix(y_test, predictions)
print(classification_report(y_test, predictions))
```

Lineer regresyon ve lojistik regresyon, iki farklı türde regresyon analizi yöntemidir ve temel olarak şu farklara sahiptirler:

Çıkış Değişkeni Türü:

Lineer Regresyon: Lineer regresyon, sürekli bir çıkış değişkenini tahmin etmek için kullanılır. Örneğin, ev fiyatlarını tahmin etmek için kullanılabilir. Çıkış değişkeni sayısal bir değerdir.
Lojistik Regresyon: Lojistik regresyon, iki kategoriden birine ait olma olasılığını tahmin etmek için kullanılır. Örneğin, bir hasta kanser taşıyıp taşımadığını tahmin etmek için kullanılabilir. Çıkış değişkeni genellikle 0 ile 1 arasında bir olasılık değeridir.
Model Tipi:

Lineer Regresyon: Lineer regresyon, bağımsız değişkenlerin doğrusal bir kombinasyonunu kullanarak çıkış değişkenini tahmin etmeye çalışır. İki veya daha fazla bağımsız değişken arasındaki ilişkiyi belirlemek için kullanılır.
Lojistik Regresyon: Lojistik regresyon, bağımsız değişkenlerin doğrusal bir kombinasyonunu sigmoid fonksiyonu içinde kullanarak çıkış değişkeni için bir olasılık tahmini yapar. Sınıflandırma problemleri için kullanılır.
Kullanım Alanları:

Lineer Regresyon: Lineer regresyon, regresyon problemlerinde kullanılır. Örneğin, gelirin yaş, eğitim seviyesi gibi faktörlere göre nasıl tahmin edilebileceğini anlamak için kullanılabilir.
Lojistik Regresyon: Lojistik regresyon, sınıflandırma problemlerinde kullanılır. Özellikle ikili sınıflandırma problemlerinde (örneğin, hasta/sağlıklı) kullanışlıdır.
Çıkış İfadesi:

Lineer Regresyon: Lineer regresyon, bir doğru denklemi şeklinde ifade edilir. Örneğin, "y = mx + b" gibi.
Lojistik Regresyon: Lojistik regresyon, log-odds (lojit) şeklinde ifade edilir. İkili sınıflandırma durumunda, "logit(p) = b0 + b1x1 + b2x2 + ... + bn*xn" gibi bir ifade kullanılır.
Hata Ölçüsü:

Lineer Regresyon: Lineer regresyonun hata ölçüsü genellikle MSE (Ortalama Kare Hatası) veya benzeri yöntemlerle hesaplanır.
Lojistik Regresyon: Lojistik regresyonun başarı ölçüsü genellikle doğruluk oranı, hassasiyet, duyarlılık gibi sınıflandırma performans metrikleri kullanılır.
Özetle, lineer regresyon sürekli bir çıkış değişkenini tahmin etmek için kullanılırken, lojistik regresyon sınıflandırma problemlerinde iki kategoriden birine ait olma olasılığını tahmin etmek için kullanılır. İki yöntem farklı problemler için tasarlanmıştır ve farklı matematiksel yapıya sahiptir.
